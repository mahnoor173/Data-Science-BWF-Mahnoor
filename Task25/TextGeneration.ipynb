# Simple Text Geenration

```python
import tensorflow as tf
import numpy as np

# Step 1: Prepare the Small Text
text = "sunrise"

# Unique characters in the text
vocab = sorted(set(text))
vocab_size = len(vocab)

# Map characters to numbers and vice versa
char2idx = {char: idx for idx, char in enumerate(vocab)}
idx2char = np.array(vocab)

# Convert the text to a sequence of numbers
text_as_int = np.array([char2idx[c] for c in text])

# Step 2: Create Training Sequences
# Input sequence (e.g., "sunris")
input_text = text_as_int[:-1]

# Target sequence (e.g., "unrise")
target_text = text_as_int[1:]

# Convert to tensors
inputs = tf.expand_dims(input_text, 0)
targets = tf.expand_dims(target_text, 0)

# Step 3: Build the Model
embedding_dim = 8
rnn_units = 16

model = tf.keras.Sequential([
    tf.keras.layers.Embedding(vocab_size, embedding_dim, input_length=len(input_text)),
    tf.keras.layers.SimpleRNN(rnn_units, return_sequences=True),
    tf.keras.layers.Dense(vocab_size)
])

# Compile the model
model.compile(optimizer='adam', loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True))

# Step 4: Train the Model
model.fit(inputs, targets, epochs=500)

# Step 5: Generate Text
def generate_text(model, start_string):
    input_eval = [char2idx[s] for s in start_string]
    input_eval = tf.expand_dims(input_eval, 0)
    text_generated = []

    for i in range(10):  # Generate 10 characters
        predictions = model(input_eval)
        predictions = tf.squeeze(predictions, 0)

        predicted_id = tf.random.categorical(predictions, num_samples=1)[-1,0].numpy()
        input_eval = tf.expand_dims([predicted_id], 0)

        text_generated.append(idx2char[predicted_id])

    return start_string + ''.join(text_generated)

# Generate new text
print(generate_text(model, start_string="s"))
```
